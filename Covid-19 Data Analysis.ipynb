##### Dropping few columns from covid_19 dataset

```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True, axis=1)
```

```python
covid_df.head()
```

##### Changing the format of the date

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

##### Find active cases

```python
covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

##### Pivot Table

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed","Deaths","Cured"], 
                           index = "State/UnionTerritory", aggfunc = max)
```

```python
#Finding out the recovery rate (totalno of cured cases/total number of confirmed cases * 100)
statewise['Recovery Rate'] = statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths']*100/statewise['Confirmed']
```

```python
# sorting values
statewise = statewise.sort_values(by = 'Confirmed', ascending = False)
```

##### Plotting a Pivot table using a nice visual

```python
statewise.style.background_gradient(cmap = "twilight")
```

## Top 10 Active Cases

##### The code first groups the DataFrame `covid_df` by the 'State/UnionTerritory' column, then applies the `max()` function to find the maximum 'Active_Cases' and corresponding 'Date' for each group. It selects only these columns, sorts the DataFrame based on 'Active_Cases' in descending order, resets the index, and stores the result in `top10_active_cases`. This DataFrame displays the states/union territories with the highest maximum active cases and their corresponding dates.

```python
top10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
```

```python
# fig size fig func
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 States with most active cases in India\n", size = 25)
```

```python
top10_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States with most active cases in India\n", size = 25)
ax = sns.barplot(data = top10_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = "black")

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show
```

## Top states with highest death Cases


```python
top10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 States with most death cases in India\n", size = 25)
ax = sns.barplot(data = top10_active_cases.iloc[:10], y = "Active_Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = "red")

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show
```

```python
#Growth Trend
fig = plt.figure(figsize = (12,6))

ax = sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x="Date", y="Active_Cases", hue='State/UnionTerritory')

ax.set_title("Top 5 Affected States in India", size = 16)
```

##### Male vs Female Vaccination

```python
male = vaccination["Male(Individuals Vaccinated)"].sum()
female = vaccination["Female(Individuals Vaccinated)"].sum()
colors = ['lightblue', 'pink']
fig = px.pie(names=["Male","Female"],values=[male,female],title = "Male and Female Vaccination",color_discrete_sequence=colors)
fig.show()
```

```python
# remove rows where state is india (as India is the country not the state false value)
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated":"Total"},inplace = True)
vaccine.head()
```

##### Most vaccinated State

```python
max_vac = vaccine.groupby(by='State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(10,5))
plt.title("Top 5 Vaccinated States in India\n", size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = "blue")

plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show
```
